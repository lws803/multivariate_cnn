```python
f = open("/Users/wilson/Desktop/NewTSCProblems/Coffee/Coffee_TEST.txt","r") # Enter your test set here
write_file = open("/Users/wilson/Desktop/NewTSCProblems/Coffee/Coffee_TEST_Combined.txt","w+") # Enter where you wish your combined outpyt to be here

lines = f.readlines()

import matplotlib.pyplot as plt

combined = []
for row in lines:
    line = row
    line = line.split()

    print("class:", int(float(line[0])), "window size:", len(line[1:]))
    line = line[1:]
    line = list(map(float, line))
    combined += list(map(float, line))
    plt.plot(line)
    plt.ylabel('some numbers')
    plt.show()


for item in combined:
    write_file.write(str(item) + '  ')

write_file.close()
f.close()

plt.plot(combined)
plt.ylabel('some numbers')
plt.show()

```
